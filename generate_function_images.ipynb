```python
import numpy as np
import pylab
import math

from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
from matplotlib.ticker import LinearLocator, FormatStrFormatter
import matplotlib.pyplot as plt


# Salomon
ff_salomon_domain = [ [-4,4], [-4,4] ];
def ff_salomon (x, y):
    r = sqrt(x*x + y*y)
    f = 1 - cos(2*pi*r) + 0.1*r
    return f

# Bohachevsky
ff_bohachevsky_domain = [ [-2, 2], [-2, 2] ]
def ff_bohachevsky (x, y):
    f = x*x + 2 *y*y - 0.3*cos(2*pi*x) - 0.4*cos(4*pi*y) + 0.7
    return f

# Griewank
ff_griewank_domain = [ [-40, 40], [-40, 40] ]
def ff_griewank (x, y):
    f = 1 + ((1/4000)*(x*x+y*y))-(cos(x)*cos(y/sqrt(2)))
    return f

# Michaelwicz
ff_michaelwicz_domain = [ [0, 4], [0, 4] ]
def ff_michaelwicz (x, y):
    a = sin(x*x / pi)
    b = sin(2*y*y / pi)
    a20 = pow(a,20)
    b20 = pow(b,20)

    f = -sin(x)*a20 - sin(y)*b20
    return f
    
def plot_contour(name, function, domain):
    numDivXY = 500
    x = np.linspace( domain[0][0], domain[0][1], numDivXY )
    y = np.linspace( domain[1][0], domain[1][1], numDivXY )
    
    X, Y = meshgrid( x, y )
    Z = function( X, Y )

    figure()
    im = imshow( Z, cmap=cm.RdBu, origin='lower', extent=[ domain[0][0], domain[0][1], domain[1][0], domain[1][1] ] )
    colorbar( im )
    
    numDivZ = 5
    scale = np.linspace( Z.min(), Z.max(), numDivZ )
    cplot = contour(X, Y, Z, scale, linewidths=2, cmap=cm.hot)
    clabel( cplot, inline=True, fmt="%1.1f", fontsize=10)
    
    title(name)
    
    savefig( name+'_contour.png', dpi=300)
    
def plot_threed(name, function, domain):
    numDivXY = 500
    x = np.linspace( domain[0][0], domain[0][1], numDivXY )
    y = np.linspace( domain[1][0], domain[1][1], numDivXY )
    
    X, Y = meshgrid( x, y )
    Z = function( X, Y )
    
    figure()
    ax = gca(projection='3d')
    surf = ax.plot_surface(X, Y, Z, rstride=1, cstride=1, 
                      cmap=cm.RdBu,linewidth=0, antialiased=False)

    ax.zaxis.set_major_locator(LinearLocator(10))
    ax.zaxis.set_major_formatter(FormatStrFormatter('%.02f'))

    colorbar(surf, shrink=0.5, aspect=5)
    
    title(name)
    
    savefig( name+'_threed.png', dpi=150)
    
def generate_plot(name, function, domain):
    plot_contour( name, function, domain )
    plot_threed( name, function, domain )
    
```

```python
generate_plot("Salomon", ff_salomon, ff_salomon_domain)
generate_plot("Bohachevsky", ff_bohachevsky, ff_bohachevsky_domain)
generate_plot("Griewank", ff_griewank, ff_griewank_domain)
generate_plot("Michaelwicz", ff_michaelwicz, ff_michaelwicz_domain)
```
